```python

def hashscrape(query):    
    # Creating list to append tweet data to
    tweets_list2 = []
    # MetaData_List = []
    tweetraw=[]
    tweetuser = []
    # Using TwitterSearchScraper to scrape data and append tweets to list
    for i,tweet in enumerate(sntwitter.TwitterHashtagScraper(query).get_items()):
        if i>100:
            break
        tweetraw.append(tweet)
        tweetuser.append(tweet.user.username)
        # print(type(tweet))
        # tweets_list2.append([tweet.date, tweet.id, tweet.rawContent, tweet.user.username,tweet.lang, tweet.likeCount, tweet.retweetCount, tweet.hashtags])
        # MetaData_List.append([tweet.lang, tweet.likeCount, tweet.retweetCount, tweet.hashtags])
    # Creating a dataframe from the tweets list above
    # tweets_df2 = pd.DataFrame(tweets_list2, columns=['Datetime', 'Tweet Id', 'Text', 'Username','Language','Likes', 'Retweets', 'Hashtags'])
    # MetaData = pd.DataFrame(MetaData_List, columns=['Language','Likes', 'Retweets', 'Hashtags'])
    # Combine lists into a dictionary
    tweetrawd = pd.DataFrame(tweetraw)
    tweetuserd = pd.DataFrame(tweetuser)
    rawdata = pd.concat([tweetrawd, tweetuserd], axis=1)
    return rawdata
```

```python
#Renaming Columns
TweetData = TweetData.rename(columns={'date': 'Datetime', 'id':'Tweet Id','rawContent': 'Text','0': 'Username', 'lang': 'Language','likeCount': 'Likes','retweetCount': 'Retweets','hashtags': 'Hashtags'})
```

```python
# Selected Data with Metadata
TweetData = TweetData.loc[:, ['Datetime', 'Tweet Id', 'Text', 'Username','Language','Likes', 'Retweets', 'Hashtags']]
TweetData
```

```python
# Total number of words in a tweet
# We considered the media data in form of a link as a word
TweetData.Text.str.split().apply(len)
```

```python
# Total number of characters in each tweet
TweetData.Text.str.len()
```

```python
# Average Word length in each tweet
def avg_word(sentence):
  words = sentence.split()
  return (sum(len(word) for word in words)/len(words))

TweetData['avg_word'] = TweetData['Text'].apply(lambda x: avg_word(x))
TweetData[['Text','avg_word']].head()
```

```python
# Lowercasing the text
TweetData['Text_Lower'] = TweetData['Text'].apply(lambda x: " ".join(x.lower() for x in x.split()))
TweetData['Text_Lower']
```

```python
# Special Characters Removal
TweetData['Text_Lower'] = TweetData['Text_Lower'].str.replace('[^\w\s]','')
TweetData['Text_Lower']
```

```python
#Number of Stopwords and Stopwords removal 
# Taking only English Language tweets
from nltk.corpus import stopwords
stop = stopwords.words('english')

Eng_Tweets = TweetData[TweetData.Language=='en'].reset_index(drop=True)
print('Number of Stopwords:', Eng_Tweets['Text_Lower'].apply(lambda x: len([x for x in x.split() if x in stop])))
Eng_Tweets['NotStopwords'] = Eng_Tweets.Text_Lower.apply(lambda x: ' '.join(x for x in x.split() if x not in stop))
print('Text with stopwords removed:')
Eng_Tweets['NotStopwords']
```

```python
# Rarewords Removal
freq = pd.Series(' '.join(Eng_Tweets['NotStopwords']).split()).value_counts()[-20:]
print('Frequency:')
print(freq)
```

```python
freql = freq.index.to_list()
Eng_Tweets['NotStopwords'] = Eng_Tweets['NotStopwords'].apply(lambda x: " ".join(x for x in x.split() if x not in freql))
Eng_Tweets['NotStopwords'].head()
```

```python
# Spelling correction
from textblob import TextBlob
Eng_Tweets['NotStopwords'][:5].apply(lambda x: str(TextBlob(x).correct()))
# This does'nt work with nouns from other languages and does'nt make sense to use for text that may contain multiple languages in a Tweet.
```

```python
# Tokenization of text
tokens = []
for x in Eng_Tweets['NotStopwords']:
    tokens.append(word_tokenize(x)) 
print('Tokens list:')
tokens
```

```python
# Stemming
stemmer = PorterStemmer()
stemmed_tokens = []
for lists in tokens:
    stemmed_tokensl = [stemmer.stem(word) for word in lists]
    stemmed_tokens.append(stemmed_tokensl)


# Lemmatization
lemmatizer = WordNetLemmatizer()
lemmatized_tokens = []
for lists in tokens:
    lemmed_tokenl = [lemmatizer.lemmatize(word) for word in lists]
    lemmatized_tokens.append(lemmed_tokenl)

print('stemmed_tokens')
stemmed_tokens
```
